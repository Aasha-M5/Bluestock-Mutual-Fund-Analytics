```python
import pandas as pd

fund_master = pd.read_csv(
    "mutual-fund-analytics/data/raw/01_fund_master.csv"
)

nav_history = pd.read_csv(
    "mutual-fund-analytics/data/raw/02_nav_history.csv"
)
```

```python
print("Fund Master Codes:",
      fund_master["amfi_code"].nunique())

print("NAV History Codes:",
      nav_history["amfi_code"].nunique())
```

```python
missing_codes = set(fund_master["amfi_code"]) - set(nav_history["amfi_code"])

print("Missing Codes:", missing_codes)
```

```python
if len(missing_codes) == 0:
    print("SUCCESS: All AMFI codes in fund_master exist in nav_history.")
else:
    print("WARNING: Missing AMFI codes found.")
```

```python
print("Fund Master Shape:", fund_master.shape)
print("NAV History Shape:", nav_history.shape)

print("\nMissing Values in Fund Master")
print(fund_master.isnull().sum())

print("\nMissing Values in NAV History")
print(nav_history.isnull().sum())
```

```python
print("Duplicate rows in Fund Master:",
      fund_master.duplicated().sum())

print("Duplicate rows in NAV History:",
      nav_history.duplicated().sum())
```

```python
import pandas as pd
import numpy as np

perf = pd.read_csv(
    "../data/processed/clean_scheme_performance.csv"
)

print(perf.columns)
print(perf.shape)
```

```python
sharpe = (
    perf[
        [
            "scheme_name",
            "sharpe_ratio"
        ]
    ]
    .sort_values(
        "sharpe_ratio",
        ascending=False
    )
)

sharpe.head(10)
```

```python
sharpe.to_csv(
    "../reports/fund_sharpe_ranks.csv",
    index=False
)

print("Saved")
```
